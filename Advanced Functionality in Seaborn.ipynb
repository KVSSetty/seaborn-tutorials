###                                 Advanced Functionality in Seaborn

Seaborn is a Python data visualization library with an emphasis on statistical plots. The library is an excellent resource for common relational,categorical and distribution plots, but where Seaborn really shines is in its ability to visualize many different features at once.

Beginners in data science may think that in a 2D plot we can have at most two variables/features in a plot depicting the relationship between those two features/variables. one variable along X-axis and the other along Y-axis. But in *seaborn* you can incorporate three more variables. So totally in a single, say a scatter plot you can have five variables realtionship depicted.

For example consider consider you have dataset made up of five variables (features) say two numeriacl and three catagorical and about 200 observations,now you can visualize the relationship between the two numerical variables by assigning one variable to x-axis another to y-axis on a scatter plot we can also visualize the dependency of the other three categorical variabes on this this realtionship. Simply assign first categorical variable to the hue attribute,second to size attribute and the third to style(shape)attribute of the scatter plot, as illustrated below with the famous but boring **tips** dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
sns.set(style="darkgrid")
```

```python
df = pd.read_csv("tips.csv")
```

```python
df.head()
```

Here in this dataset we have two numeric variables **total_bill** and **tip** five categorical variables **sex,smoker day,time** and **size**.

First let us plot a simple scatter plot to visualize the relationship of tip with total-bill

```python
sns.scatterplot(x="total_bill",y="tip",data=df)
```

Now say you want know how this relationship depend on the tip givers sex,we can easily do this by assigning the sex categorical variable to the hue attribute of the scatter plot as below:

```python
sns.scatterplot(x="total_bill",y="tip",hue="sex",data=df)
```

Very similarly we can visualize how the relationship depends on any two other variables, say on time ie launch or dinner and size of the table as follows

```python
sns.relplot(x="total_bill",y="tip",hue="sex",style="time",size="size",data=df)
```

#### Showing multiple relationships with facets

Weve emphasized in the previous section that, while these functions can show several semantic variables at once, its not always effective to do so. But what about when you do want to understand how a relationship between two variables depends on more than one other variable?

Say you want present the **tip** and **total-bill** relationship depends on **sex,time** and **somker** in simple, understandableand easily digestable by even a common person. 

The best approach may be to make more than one plot. Because relplot() is based on the FacetGrid, this is easy to do. To show the influence of an additional variable, instead of assigning it to one of the semantic roles(like style or size) in the plot, use it to facet the visualization. This means that you make multiple axes and plot subsets of the data on each of them:

```python
sns.relplot(x="total_bill", y="tip", hue="sex",
            col="time",row="smoker", data=df)
```

Let me give another example this time say you want know how the table size influences the tip:

```python
sns.relplot(x="total_bill", y="tip", hue="sex",
            col="time",row="size",height=4,aspect=1, data=df)
```

These visualizations, which are often called lattice plots or small-multiples, are very effective because they present the data in a format that makes it easy for the eye to detect both overall patterns and deviations from those patterns. While you should make use of the flexiblity afforded by scatterplot() and relplot(), always try to keep in mind that several simple plots are usually more effective than one complex plot.
